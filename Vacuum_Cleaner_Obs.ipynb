- Evironment: Môi trường agent làm việc - Sàn và bụi

- Actuators: Bộ truyền động/ Hành động Agents có thể làm - Hút bụi, di chuyển (lên xuống trái phải)

- Sensors: Các cách để Agents cảm nhận môi trường - Các cảm biến (Vị trí, nhận diện bụi, nhận diện vật cản)

# Rules

    - if x > 0: move.append("DOWN")

    - if x < Row - 1: move.append("UP")

    - if y > 0: move.append("RIGHT")

    - if y < COL - 1: move.append("LEFT")

# Obstacles (Vật cản)

- Vật cản chiếm một ô trên sàn và máy hút bụi không thể đi vào ô đó.
-  Trong hàm `floor_initialize`, sàn nhà sẽ được tạo ngẫu nhiên với tỷ lệ xác định cho ô sạch (`0`), ô có bụi (`1`), và vật cản (`-1`). Sẽ không sinh ô có vật cản ở vị trí bắt đầu của máy
-  Hàm `posible_move` xét ô sàn muốn di chuyển (`floor[x][y] != -1`). Nếu là vật cản, hướng đi đó sẽ không được thêm vào danh sách các bước di chuyển khả thi.
-  Khi 4 phía di chuyển đều bị chặn, trạng thái (`Vacuum is stuck!`) được in và tự động kết thúc sớm để tránh lặp vô hạn.


```python
import random

ROW = random.randint(3, 5)
COL = random.randint(3, 5)

vacuum_pos = (random.randint(0, ROW-1), random.randint(0, COL-1))

def floor_initialize(x, y):
    """
    This function return the floor with random dirty tiles and obstacles (-1)
    """
    floor = []
    for i in range(x):
        # 0: Sạch, 1: Bụi, -1: Vật cản. Xác suất vật cản 15%
        floor.append([random.choices([0, 1, -1], weights=[0.5, 0.35, 0.15])[0] for j in range(y)])
    
    # Đảm bảo vị trí ban đầu của máy hút bụi không phải là vật cản
    if floor[vacuum_pos[0]][vacuum_pos[1]] == -1:
        floor[vacuum_pos[0]][vacuum_pos[1]] = 0
        
    return floor

def posible_move(x, y, floor):
    """
    This function return the list of possible moves avoiding obstacles
    """
    move = []
    if (x > 0) and floor[x-1][y] != -1:
        move.append("UP")
    if (x < ROW - 1) and floor[x+1][y] != -1:
        move.append("DOWN")
    if (y > 0) and floor[x][y-1] != -1:
        move.append("LEFT")
    if (y < COL - 1) and floor[x][y+1] != -1:
        move.append("RIGHT")
    return move

def is_clean(inner_floor):
    """
    This function return True if the floor is clean
    """
    for i in range(ROW):
        for j in range(COL):
            if (inner_floor[i][j] == 1):
                return False
    return True

def main():
    global vacuum_pos
    
    floor = floor_initialize(ROW, COL)
    move_count = 0 #đếm số lần di chuyển
    
    print(f"Floor: {floor}")
    print(f"Vacuum Position: {vacuum_pos}")
    
    #dừng khi sàn sạch hoặc move_count lớn hơn 20
    while (not is_clean(floor) and move_count < 20):

        move = posible_move(vacuum_pos[0], vacuum_pos[1], floor)
        
        if (move):
            move = random.choice(move) #random hướng di chuyển trong move
            if (move == "UP"):
                vacuum_pos = (vacuum_pos[0] - 1, vacuum_pos[1])
            elif (move == "DOWN"):
                vacuum_pos = (vacuum_pos[0] + 1, vacuum_pos[1])
            elif (move == "LEFT"):
                vacuum_pos = (vacuum_pos[0], vacuum_pos[1] - 1)
            elif (move == "RIGHT"):
                vacuum_pos = (vacuum_pos[0], vacuum_pos[1] + 1)
            
            print(f"Vacuum moved to position {vacuum_pos}")
            print("=" * 20)
            
            if floor[vacuum_pos[0]][vacuum_pos[1]] != -1:
                floor[vacuum_pos[0]][vacuum_pos[1]] = 0
        else:
            print("Vacuum is stuck! (No possible moves)")
            break
        
        move_count += 1
        print(f"Floor: {floor}")
        print(f"Vacuum Position: {vacuum_pos}")
    

    #in
    if (is_clean(floor)):
        print("Floor is clean")
    else:
        print("Floor is not clean")
        
    print(f"Total moves: {move_count}")

if (__name__ == "__main__"):
    main()
```
