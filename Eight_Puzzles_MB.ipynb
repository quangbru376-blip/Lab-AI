```python
import random
import copy

GOAL_STATE = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]

def matrix_to_tuple(matrix):
    """
    Chuyển từ list sang tuple
    """
    return tuple(tuple(row) for row in matrix)

def find_blank(board):
    """
    Tìm vị trí của số 0
    """
    for r in range(3):
        for c in range(3):
            if board[r][c] == 0:
                return (r, c)
    return None

def get_valid_moves(r, c):
    """
    Lấy danh sách các bước di chuyển hợp lệ
    """
    moves = []
    if r > 0: moves.append("UP")
    if r < 2: moves.append("DOWN")
    if c > 0: moves.append("LEFT")
    if c < 2: moves.append("RIGHT")
    return moves

def apply_move(board, move):
    """
    Áp dụng bước di chuyển
    """
    r, c = find_blank(board)
    new_board = copy.deepcopy(board)
    if move == "UP":
        new_board[r][c], new_board[r-1][c] = new_board[r-1][c], new_board[r][c]
    elif move == "DOWN":
        new_board[r][c], new_board[r+1][c] = new_board[r+1][c], new_board[r][c]
    elif move == "LEFT":
        new_board[r][c], new_board[r][c-1] = new_board[r][c-1], new_board[r][c]
    elif move == "RIGHT":
        new_board[r][c], new_board[r][c+1] = new_board[r][c+1], new_board[r][c]
    return new_board

def generate_random_board():
    """
    Tạo ra một bàn cờ ngẫu nhiên
    """
    numbers = list(range(9))
    
    random.shuffle(numbers)
    
    board = [numbers[i:i+3] for i in range(0, 9, 3)]
    
    return board

def check_if_visited(board, moves):
    """
    Kiểm tra xem các bước di chuyển có dẫn đến trạng thái đã có rồi không
    Nếu có thì loại bỏ khỏi danh sách các bước di chuyển hợp lệ
    """
    state_tuple = matrix_to_tuple(board)
    for move in moves:
        temp_board = apply_move(board, move)
        temp_state_tuple = matrix_to_tuple(temp_board)
        if temp_state_tuple in visited_states:
            moves.remove(move)
    return moves 



current_board = generate_random_board()
visited_states = set()
step = 0

def main():
    global step, visited_states, current_board

    print("Trạng thái ban đầu:")
    for i in current_board:
        print(i)
    
    while True:
        step += 1
        state_tuple = matrix_to_tuple(current_board)
        
        if current_board == GOAL_STATE:
            print(f"Đạt đích sau {step} bước")
            break
        
        r, c = find_blank(current_board)
        valid_moves = get_valid_moves(r, c)
        valid_moves = check_if_visited(current_board, valid_moves)

        if valid_moves == []:
            print("Không tìm thấy bước đi hợp lệ")
            break

        action = random.choice(valid_moves)
        current_board = apply_move(current_board, action)
        visited_states.add(state_tuple)
        
        #In bàn cờ
        print("=" * 20)
        print(f"Bước {step}: Di chuyển {action}")
        for i in current_board:
            print(i)
        
        if step > 20:
            print("Dừng do quá nhiều bước (Trên 20)")
            break

if __name__ == "__main__":
    main()
```
