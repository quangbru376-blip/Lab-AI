```python
import random


ROW = 4
COL = 4
vacuum_pos = (random.randint(0, ROW-1), random.randint(0, COL-1))
vacuum_pos = (2,  3)

def matrix_to_tuple(matrix):
    """
    This function convert the matrix to the tuple. The main reason is to save the mem
    """
    return tuple(map(tuple, matrix))

def floor_initialize(x, y):
    """
    This function return the floor with random dirty tiles and obstacles (-1)
    """
    floor = []
    for i in range(x):
        # 0: Sạch, 1: Bụi, -1: Vật cản. Xác suất vật cản 15%
        floor.append([random.choices([0, 1, -1], weights=[0.5, 0.35, 0.15])[0] for j in range(y)])
    
    # Đảm bảo vị trí ban đầu của máy hút bụi không phải là vật cản
    if floor[vacuum_pos[0]][vacuum_pos[1]] == -1:
        floor[vacuum_pos[0]][vacuum_pos[1]] = 0
        
    return floor

def posible_move(x, y, floor):
    """
    This function return the list of possible moves avoiding obstacles
    """
    moves = []
    if (x > 0) and floor[x-1][y] != -1:
        moves.append("UP")
    if (x < ROW - 1) and floor[x+1][y] != -1:
        moves.append("DOWN")
    if (y > 0) and floor[x][y-1] != -1:
        moves.append("LEFT")
    if (y < COL - 1) and floor[x][y+1] != -1:
        moves.append("RIGHT")
    return moves

def is_clean(floor):
    """
    This function return True if the floor is clean
    """
    for i in range(ROW):
        for j in range(COL):
            if (floor[i][j] == 1):
                return False
    return True

def apply_move(floor, vacuum_pos, move):
    if (move == "UP"):
        vacuum_pos = (vacuum_pos[0] - 1, vacuum_pos[1])
    elif (move == "DOWN"):
        vacuum_pos = (vacuum_pos[0] + 1, vacuum_pos[1])
    elif (move == "LEFT"):
        vacuum_pos = (vacuum_pos[0], vacuum_pos[1] - 1)
    elif (move == "RIGHT"):
        vacuum_pos = (vacuum_pos[0], vacuum_pos[1] + 1)

    if floor[vacuum_pos[0]][vacuum_pos[1]] != -1:
        floor[vacuum_pos[0]][vacuum_pos[1]] = 0
        update_visited_state(floor, vacuum_pos)
    
    return floor, vacuum_pos

def update_visited_state(floor, vacuum_pos):
    """
    This function storage all the previous floor state and vacuum position
    """
    global visited_states
    tup = matrix_to_tuple(floor)
    state = (tup, vacuum_pos)
    if state not in visited_states:
        visited_states.append(state)

def check_if_visited(floor, vacuum_pos):
    moves = posible_move(vacuum_pos[0], vacuum_pos[1], floor)
    if (not moves):
        return []
    
    valid_moves = []
    for move in moves:
        temp_floor = [row[:] for row in floor]
        temp_pos = vacuum_pos
        
        if (move == "UP"):
            temp_pos = (temp_pos[0] - 1, temp_pos[1])
        elif (move == "DOWN"):
            temp_pos = (temp_pos[0] + 1, temp_pos[1])
        elif (move == "LEFT"):
            temp_pos = (temp_pos[0], temp_pos[1] - 1)
        elif (move == "RIGHT"):
            temp_pos = (temp_pos[0], temp_pos[1] + 1)

        if temp_floor[temp_pos[0]][temp_pos[1]] != -1:
            temp_floor[temp_pos[0]][temp_pos[1]] = 0
            
        temp_tuple_state = matrix_to_tuple(temp_floor)

        if (temp_tuple_state, temp_pos) not in visited_states:
            valid_moves.append(move)
            
    return valid_moves


floor = [[-1, 0, 1, 0],
         [0, 0, -1, 0],
         [1, 0, -1, 0],
         [0, 0, 1, 0]]
visited_states = []

def main():
    global vacuum_pos, floor
    move_count = 0 #đếm số lần di chuyển
    
    # dọn nếu vị trí bắt đầu cũng bẩn
    if floor[vacuum_pos[0]][vacuum_pos[1]] != -1:
        floor[vacuum_pos[0]][vacuum_pos[1]] = 0
    update_visited_state(floor, vacuum_pos)
    
    print("Initial floor: ")
    for i in floor:
        print(i)
    print(f"Vacuum Position: {vacuum_pos}")
    
    #dừng khi sàn sạch hoặc move_count lớn hơn 20
    while (not is_clean(floor) and move_count < 20):

        moves = check_if_visited(floor, vacuum_pos)
        
        if (moves):
            move = random.choice(moves) #random hướng di chuyển trong move
            floor, vacuum_pos = apply_move(floor, vacuum_pos, move)
            
            print("=" * 20)

        else:
            print("No posible move")
            break
        
        move_count += 1
        print("Floor: ")
        for i in floor:
            print(i)
        print(f"Vacuum Position: {vacuum_pos}")
    

    #in
    if (is_clean(floor)):
        print("Floor is clean")
    else:
        print("Floor is not clean")
        
    print(f"Total moves: {move_count}")

if (__name__ == "__main__"):
    main()
```
