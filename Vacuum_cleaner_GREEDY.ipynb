# Vacuum Cleaner - Greedy (Ưu tiên số ô bẩn ít nhất - gần với GOAL nhất)

## function Greedy_Search(Start, Goal):
- `h(n) = min(number_of_dirty_tiles(i) for i in frontier)`
1. Khởi tạo tập FRONTIER = {Start}

    Tính hàm đánh giá Start: h(Start)

2. Khởi tạo tập REACHED = {}

3. TRONG KHI (FRONTIER không rỗng):

- a. Chọn trạng thái n từ FRONTIER có h(n) nhỏ nhất.

- b. NẾU n == Goal:

    TRẢ VỀ "Thành công" và truy xuất lại đường đi từ Start đến n.
        
- c. Loại bỏ n khỏi FRONTIER và thêm n vào REACHED.

    - d. Với mỗi trạng thái m kề với n:

    i. NẾU m chưa có trong cả FRONTIER và REACHED:

    - Gán đỉnh cha của m là n.

    - Tính giá trị heuristic h(m).

    - Thêm m vào FRONTIER.

    ii. NẾU m đã có trong FRONTIER hoặc REACHED:

    - Bỏ qua m.

4. TRẢ VỀ "Thất bại" (Không tìm thấy đường đi).
