# Rules
- posible.move(x, y)

    - if x > 0: move.append("DOWN")

    - if x < Row - 1: move.append("UP")

    - if y > 0: move.append("RIGHT")

    - if y < COL - 1: move.append("LEFT")

- if state.value = 0 then action = random(move)
- if state.value = 1 then state.value = 0 and action = random(move)

```python
import random

ROW = random.randint(2, 5)
COL = random.randint(2, 5)

vacuum_pos = (random.randint(0, ROW-1), random.randint(0, COL-1))

def floor_initialize(x, y):
    """
    This function return the floor with random dirty tiles
    """
    floor = []
    for i in range(x):
        floor.append([random.randint(0, 1) for j in range(y)])
    return floor

def posible_move(x, y):
    """
    This function return the list of possible moves
    """
    move = []
    if (x > 0):
        move.append("UP")
    if (x < ROW - 1):
        move.append("DOWN")
    if (y > 0):
        move.append("LEFT")
    if (y < COL - 1):
        move.append("RIGHT")
    return move

def is_clean(inner_floor):
    """
    This function return True if the floor is clean
    """
    for i in range(ROW):
        for j in range(COL):
            if (inner_floor[i][j] == 1):
                return False
    return True

def main():
    global vacuum_pos
    
    floor = floor_initialize(ROW, COL)
    move_count = 0 #đếm số lần di chuyển
    
    print(f"Floor: {floor}")
    print(f"Vacuum Position: {vacuum_pos}")
    
    #dừng khi sàn sạch hoặc move_count lớn hơn 100
    while (not is_clean(floor) and move_count < 20):

        move = posible_move(vacuum_pos[0], vacuum_pos[1])
        
        if (move):
            move = random.choice(move) #random hướng di chuyển trong move
            if (move == "UP"):
                vacuum_pos = (vacuum_pos[0] - 1, vacuum_pos[1])
            elif (move == "DOWN"):
                vacuum_pos = (vacuum_pos[0] + 1, vacuum_pos[1])
            elif (move == "LEFT"):
                vacuum_pos = (vacuum_pos[0], vacuum_pos[1] - 1)
            elif (move == "RIGHT"):
                vacuum_pos = (vacuum_pos[0], vacuum_pos[1] + 1)
            
            print(f"Vacuum moved to position {vacuum_pos}")
            print("=" * 20)
        
        floor[vacuum_pos[0]][vacuum_pos[1]] = 0
        move_count += 1
        print(f"Floor: {floor}")
        print(f"Vacuum Position: {vacuum_pos}")
    

    #in
    if (is_clean(floor)):
        print("Floor is clean")
    else:
        print("Floor is not clean")
        
    print(f"Total moves: {move_count}")

if (__name__ == "__main__"):
    main()
```
